```python
import torch
import collections
```

```python
state_dict = torch.load('ckpt.pt')
```

```python
for name, param in state_dict.items():
    print(f"Layer/Module: {name}")
    
    if isinstance(param, torch.Tensor):  # If param is a tensor
        print(f"Shape: {param.shape}\n")
        
    elif isinstance(param, collections.OrderedDict):  # If param is an OrderedDict
        print("============================")
        for sub_name, sub_param in param.items():
            if isinstance(sub_param, torch.Tensor):  # If the item inside OrderedDict is a tensor
                print(f"Sub-Layer: {sub_name}")
                print(f"Shape: {sub_param.shape}\n")
                
    else:
        print(f"Object is of type: {type(param)}\n")
        
config = state_dict['config']
print(config
      
```
